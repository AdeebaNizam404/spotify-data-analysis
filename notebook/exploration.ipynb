Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading dataset

```python
df = pd.read_csv("../data/Spotify.csv")
```

Dataset Overview

```python
print(df.shape)
print(df.columns)
df.head()
df.tail()
```

Basic info

Correcting the  minor missing values

```python
df['duration_min'] = df['duration_ms'] / 60000
```

Categorical Exploration

```python
print(df['artists'].nunique())  #number of unique artists
print(df['album_name'].nunique())  #number of unique albums
print(df['track_name'].nunique())  #number of unique tracks
df['artists'].value_counts().head(10)  #top artist by number of songs
```

Numerical Exploration

```python
df['popularity'].hist(bins=30,figsize=(8,5))
plt.title('song Popularity Distribution')
plt.xlabel("Popularity")
plt.ylabel("count")
plt.show()
#Average duration
df['duration_min']=df['duration_ms']/60000
df['duration_min'].describe()
```

Correlation Check

```python
from unicodedata import numeric

numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),cmap="YlGnBu",annot=True)
plt.title("correlation Heatmap of Numerical features")
plt.show()
```
